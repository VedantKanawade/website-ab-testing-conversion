```python
import pandas as pd

primary = pd.read_csv("../data/processed/conversion_results.csv")
secondary = pd.read_csv("../data/processed/secondary_results.csv")
```

```python
summary = pd.concat([primary, secondary], ignore_index=True)
summary
```

```python
summary["decision"] = "hold"

# Primary metric
if primary.loc[0,"p_value"] < 0.05 and primary.loc[0,"uplift"] > 0:
    summary.loc[0,"decision"] = "ship"

# Secondary metrics check
for i in range(1, len(summary)):
    if summary.loc[i,"metric"] == "bounce_rate" and summary.loc[i,"uplift"] > 0.005:
        summary.loc[0,"decision"] = "iterate"
    if summary.loc[i,"metric"] == "avg_revenue" and summary.loc[i,"uplift"] < -0.01:
        summary.loc[0,"decision"] = "iterate"
```

```python
summary.to_csv("../reports/experiment_summary.csv", index=False)
```
